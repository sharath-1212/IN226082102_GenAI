## Task 1: Conceptual Understanding

**1. Difference between "Love" and "love" in NLP:**
In NLP, “Love” and “love” are treated as different tokens due to case sensitivity. This can lead to duplicate representations of the same word, increasing vocabulary size and reducing model efficiency. Converting text to lowercase helps normalize data, ensuring consistency and better learning for machine learning models.

**2. What happens if stopwords are not removed?**
If stopwords are not removed, the dataset contains many common words like “the,” “is,” and “and,” which add noise and do not contribute much meaning. This can increase computational cost, reduce model accuracy, and make it harder for algorithms to focus on important, meaningful words.

**3. Two real-world scenarios where removing stopwords can be harmful:**
In sentiment analysis, removing words like “not” can completely change meaning, such as “not good” becoming “good.” In question-answering systems, removing words like “what” or “how” can distort the intent of the query, making it difficult for the system to understand user questions correctly.

**4. Difference between stemming and lemmatization:**
Stemming reduces words to their base form by cutting off suffixes, often resulting in incomplete or non-dictionary words like “running” to “run.” Lemmatization, however, uses vocabulary and linguistic rules to convert words into their correct base form, ensuring meaningful and accurate root words such as “better” to “good.”


## Task 2: Advanced Preprocessing Function

```python

import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Remove URLs & emails
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters/emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short tokens except 'no', 'not'
    tokens = [t for t in tokens if len(t) > 2 or t in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

## Task 3: Stress Testing

```python

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://amazon.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in samples:
    tokens, clean = preprocess_text(text)
    results.append((text, tokens, clean))
    print("Original:", text)
    print("Tokens:", tokens)
    print("Clean:", clean)
    print("-"*40)
```

## Task 4: Token Analytics

```python

analytics = []

for text, tokens, clean in results:
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens)/total if total > 0 else 0

    analytics.append((text, total, unique, avg_len))
    print(f"Sentence: {text}")
    print(f"Total Tokens: {total}, Unique: {unique}, Avg Length: {avg_len:.2f}")
    print("-"*40)

print("\nAnalysis:")
print("Most noisy sentence: Likely one with symbols/numbers like 'Win $$$ now!!!'")
print("Most meaningful: 'I am not happy with this'")
```

## Task 5: Frequency Analysis

```python

from collections import Counter

all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Top 5 least frequent:", counter.most_common()[-5:])
```

## Task 6: Full Pipeline

```python

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example
pipeline_output = full_pipeline(samples)
print(pipeline_output)
```

## Task 7: Error Handling

```python
test_cases = [
    "",                 # Empty string
    "😍😍😂😂🔥🔥",     # Only emojis
    "1234567890",       # Only numbers
    "!!!@@@",           # Only symbols
    "   ",              # Only spaces
    None                # None input
]

for i, test in enumerate(test_cases, 1):
    tokens, cleaned = preprocess_text(test)
    print(f"\nTest Case {i}:")
    print("Input            :", test)
    print("Tokens           :", tokens)
    print("Cleaned Sentence :", cleaned)
    print("-" * 50)
```
